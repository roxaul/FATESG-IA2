# Exercicio
Criar um programa em Python que recebe um nome completo de uma pessoa e:
Conta quantos caracteres tem cada nome.
Identifica a posição de cada nome dentro do nome completo.
Reescreve cada nome de trás para frente.

```python
nome_completo = input("Digite seu nome completo: ")
nomes = nome_completo.split()

print("-" * 30)
for i, nome in enumerate(nomes):
    print(f"Nome: {nome}")
    print(f"Quantidade de caracteres: {len(nome)}")
    print(f"Posição: {i + 1}º nome")
    print(f"Invertido: {nome[::-1]}")
    print("-" * 30)
```
