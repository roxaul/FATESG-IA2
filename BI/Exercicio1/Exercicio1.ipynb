## Parte 1: Numpy

### Exercício 1.1: Criação de Arrays

## Parte 2: Pandas

### Exercício 2.1: Criação de DataFrame
Crie um DataFrame chamado `df` a partir do dicionário abaixo.

```python
import pandas as pd

dados = {
    'Nome': ['Alice', 'Bob', 'Charlie', 'David'],
    'Idade': [25, 30, 35, 40],
    'Cidade': ['São Paulo', 'Rio de Janeiro', 'Belo Horizonte', 'Curitiba']
}

# Seu código aqui
df = pd.DataFrame(dados)
print(df)
```

### Exercício 2.2: Filtragem de Dados
Filtre o DataFrame `df` para mostrar apenas as pessoas com mais de 30 anos.

```python
# Seu código aqui
maiores_30 = df[df['Idade'] > 30]
print(maiores_30)
```
