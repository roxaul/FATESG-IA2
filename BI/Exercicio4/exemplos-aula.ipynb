```python
import pandas as pd
#Usando o DataFrame anterior, vamos deletar uma coluna.
#Usando a função del
d = {"Um": pd.Series([1,2,3], index = ["A","B","C"]), "Dois": pd.Series([1,2,3,4], index = ["A","B","C","D"]), "Três": pd.Series([10,20,30], index = ["A","B","C"]),}

df = pd.DataFrame(d)
print("O dataframe é: ")
print(df)
```

```python
#Usando a função del
print("Excluindo a primeira coluna usando a função DEL: ")
del df["Um"]
print(df)
```

```python
import pandas as pd

# Reutilizando o dicionário da imagem anterior
data = {
    'Estado': ['Santa Catarina', 'Paraná', 'Goiás', 'Bahia', 'Minas Gerais'],
    'Ano': [2002, 2003, 2004, 2005, 2006],
    'População': [1.5, 1.7, 3.6, 2.4, 2.9]
}

# Criando o DataFrame data2 com colunas específicas e índices personalizados
data2 = pd.DataFrame(data, columns=['Ano', 'Estado', 'População', 'Débito'],
                     index=['um', 'dois', 'três', 'quatro', 'cinco'])

print(data2)
```

```python
import numpy as np

# Atribuindo valores de 0 a 4 (sequência de 5 números) para a coluna Débito
data2['Débito'] = np.arange(5.) #inserir valores entre 0 e 5.

print(data2)
```

```python
import pandas as pd
import numpy as np

# Criando um dicionário de séries
d = {'Name':pd.Series(['Tom','James','Ricky','Vin','Steve','Smith','Jack']),
     'Idade':pd.Series([25,26,25,23,30,29,23]),
     'Avaliação':pd.Series([4.23,3.24,3.98,2.56,3.20,4.6,3.8])}

# Criando o dataframe
df = pd.DataFrame(d)

print(df)
print(60*"#")
#T: Retorna a transposição do DataFrame. As linhas e colunas serão trocadas.
print(df.T)
print(60*"#")
#axes: Retorna uma lista com os rótulos da linha e da coluna como os únicos membros.
print(df.axes)
print(60*"#")
#dtypes: Retorna o tipo de dados de cada coluna
print(df.dtypes)
print(60*"#")
#empty: Retorna o valor booleano informando se o Objeto está vazio ou não; True indica que o objetoestá vazio.
print(df.empty)
print(60*"#")
#ndim: Retorna o número de dimensões do objeto. Por definição, DataFrame é um objeto 2D.
print(df.ndim)
```
