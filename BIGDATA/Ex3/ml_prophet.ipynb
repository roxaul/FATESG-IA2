```python
#Importação das Libs
import pandas as pd
import prophet as ph
```

```python
data = pd.read_csv('casos-covid-go.csv')
data.tail()
```

```python
data.drop(["date","state","city","place_type","is_last"], axis = 1).corr()
```

```python
estados = data.loc[data.place_type == 'state']
estados_go = estados.loc[data.state == 'GO']
estados_go_casos = estados_go.groupby('date')['confirmed'].sum().reset_index()
estados_go_casos.tail()
```

```python
estados_go_casos_previsao = estados_go_casos
estados_go_casos_previsao.columns = ['ds','y']
estados_go_casos_previsao.tail()
```

```python
# Criar instância do Prophet
modelProphet = ph.Prophet()
# Treinamento do modelos
modelProphet.fit(estados_go_casos_previsao)
# Definição de futuro para a realização da previsão
futuro = modelProphet.make_future_dataframe(30)
# Previsão dentro de um cenário futuro
previsao = modelProphet.predict(futuro)
# Plotando o resultado
fig = modelProphet.plot(previsao)
```

```python
cidade_goiania_mortes = data.loc[data.place_type=='city'].loc[data.city=='Goiânia'].groupby('date')['deaths'].sum().reset_index()
cidade_goiania_mortes.columns = ['ds','y']

modelProphet = ph.Prophet()
modelProphet.fit(cidade_goiania_mortes)
futuro = modelProphet.make_future_dataframe(60)
previsao = modelProphet.predict(futuro)
fig = modelProphet.plot(previsao)
```
