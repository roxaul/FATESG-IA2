```python
import pandas as pd
import matplotlib.pyplot as plt
# Carregar o dataset caso ainda não esteja carregado
df = pd.read_csv('Smartphone_Usage_And_Addiction_Analysis_7500_Rows (1).csv')
```

```python
# 1. Média de idade
media_idade = df['age'].mean()
print(f"Média de idade: {media_idade:.2f}")
```

```python
# 2. Média de horas de sono
media_sono = df['sleep_hours'].mean()
print(f"Média de horas de sono: {media_sono:.2f}")
```

```python
# 3. Mínimo de horas de sono
min_sono = df['sleep_hours'].min()
print(f"Mínimo de horas de sono: {min_sono}")
```

```python
# 4. Máximo de horas em rede social
max_rede_social = df['social_media_hours'].max()
print(f"Máximo de horas em rede social: {max_rede_social}")
```

```python
# 5. Describe das horas de jogos (gaming_hours)
print("\n--- Estatísticas descritivas (describe) de gaming_hours ---")
print(df['gaming_hours'].describe())
```

```python
# pd.crosstab cruza as idades com os níveis de vício e faz a contagem de pessoas
tabela_cruzada = pd.crosstab(df['age'], df['addiction_level'])
# Gera o gráfico de barras
tabela_cruzada.plot(
    kind='bar', 
    figsize=(14, 7), 
    width=0.8, # Deixa as barras levemente mais grossas
    cmap='viridis' # Define as cores
)
# Configurações do gráfico
plt.title('Frequência de Nível de Vício distribuída por Idades', fontsize=14)
plt.xlabel('Idade', fontsize=12)
plt.ylabel('Quantidade de Usuários', fontsize=12)
# Coloca a legenda fora do gráfico para não tampar as barras
plt.legend(title='Nível de Vício', bbox_to_anchor=(1.05, 1), loc='upper left')
# Mantém as idades retas no Eixo X
plt.xticks(rotation=0)
# Ajustar layout para a legenda não cortar
plt.tight_layout()
plt.show()
```

```python
 #Conta a quantidade de 'user_id' únicos para cada categoria de 'stress_level'
contagem_stress = df.groupby('stress_level')['user_id'].count()
# Cria o gráfico de pizza a partir dessa contagem
contagem_stress.plot(
    kind='pie',               # Define que é gráfico de pizza (torta)
    figsize=(8, 8),           # Tamanho do gráfico
    autopct='%1.1f%%',        # Formata para exibir as porcentagens (Ex: 33.5%)
    startangle=140,           # Gira o gráfico para começar de um ângulo melhor
    colors=['#ff9999', '#66b3ff', '#99ff99','#ffcc99'], # Cores suaves 
    textprops={'fontsize': 12}
)
# Configurações de layout
plt.title('Distribuição de Usuários por Nível de Estresse', fontsize=16, fontweight='bold')
plt.ylabel('')  # Remove a legenda do eixo Y (no caso da pizza, atrapalha a estética)
plt.show()
```
